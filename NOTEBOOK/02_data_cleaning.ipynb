```python
import pandas as pd
import numpy as np
#导入健康数据汇总csv
df=pd.read_csv('D:/DA/Huawei_Sleep_Analysis/DATA/processed/health_daily_summary.csv')
#预览数据前五行
df.head()
```

```python
df.isnull().sum()
```

```python
#用各列的平均值填充缺失值
df['REM时长']=df['REM时长'].fillna(df['REM时长'].mean())

df['午睡时长']=df['午睡时长'].fillna(df['午睡时长'].mean())

df['深睡时长']=df['深睡时长'].fillna(df['深睡时长'].mean())

df['浅睡时长']=df['浅睡时长'].fillna(df['浅睡时长'].mean())

df['睡眠效率']=df['睡眠效率'].fillna(df['睡眠效率'].mean())

df['清醒时长']=df['清醒时长'].fillna(df['清醒时长'].mean())

df['夜间总睡眠']=df['夜间总睡眠'].fillna(df['夜间总睡眠'].mean())

df['深睡比例']=df['深睡比例'].fillna(df['深睡比例'].mean())

df['浅睡比例']=df['浅睡比例'].fillna(df['浅睡比例'].mean())

df['快速眼动比例']=df['快速眼动比例'].fillna(df['快速眼动比例'].mean())

df['在床时间']=df['在床时间'].fillna(df['在床时间'].mean())

df['总时长']=df['总时长'].fillna(df['总时长'].mean())

df['总步数']=df['总步数'].fillna(df['总步数'].mean())

df['总消耗']=df['总消耗'].fillna(df['总消耗'].mean())

df['每公里消耗']=df['每公里消耗'].fillna(df['每公里消耗'].mean())

#用该列第一个众数填充缺失值
df['是否午睡']=df['是否午睡'].fillna(df['是否午睡'].mode()[0])

df['星期']=df['星期'].fillna(df['星期'].mode()[0])

df.isnull().sum()
```

```python
#查询重复列
dup=df.duplicated().sum()
print(dup)
```

```python
#查看各字段基础统计值，查看极端值是否合理
df.describe()
```

```python
#处理异常值
df['REM时长']=df['REM时长'].replace(0,np.nan)

df['浅睡时长']=df['浅睡时长'].replace(0,np.nan)

df['深睡时长']=df['深睡时长'].replace(0,np.nan)

df['夜间总睡眠']=df['夜间总睡眠'].replace(0,np.nan)

df['在床时间']=df['在床时间'].replace(0,np.nan)

#再次查看基础统计值
df.describe()
```
